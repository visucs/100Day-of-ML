#**EDA using Bivariate and Multivariate Analysis**

1Ô∏è‚É£ What is EDA?

EDA (Exploratory Data Analysis) is the process of analyzing datasets to understand:

Data distribution

Relationships between variables

Patterns & trends

Outliers

Feature importance

EDA helps before building ML models.

2Ô∏è‚É£ Types of EDA

Univariate Analysis ‚Üí 1 variable

Bivariate Analysis ‚Üí 2 variables

Multivariate Analysis ‚Üí 3+ variables

üîµ 3Ô∏è‚É£ Bivariate Analysis (2 Variables)

üëâ Used to find relationship between two variables.

üìå Case 1: Numerical vs Numerical

Example: Study Hours vs Marks

üéØ Goal:

Check if increasing study hours increases marks

üìå Interpretation:

Positive correlation ‚Üí marks increase with study hours

Negative correlation ‚Üí opposite relation

üìå Case 2: Categorical vs Numerical

Example: Gender vs Salary

üìå Interpretation:

Compare median salary

Check salary spread

üìå Case 3: Categorical vs Categorical

Example: Gender vs Purchased


```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
df.head()
```

#1. Categorical Data

#a. Countplot

```python
sns.countplot(df['Survived'])
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(df['Pclass'])
df['Pclass'].value_counts().plot(kind='bar')
```

#b. PieChart

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```
